```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Better looking plots
plt.style.use("ggplot")

# Load Dataset
df = pd.read_csv("../data/StudentPerformanceFactors.csv")

# ==========================
# BASIC DATA ANALYSIS
# ==========================

print("="*50)
print("FIRST 5 ROWS")
print("="*50)
print(df.head())

print("\n")

print("="*50)
print("LAST 5 ROWS")
print("="*50)
print(df.tail())

print("\n")

print("="*50)
print("DATASET SHAPE")
print("="*50)
print(df.shape)

print("\n")

print("="*50)
print("DATASET INFORMATION")
print("="*50)
print(df.info())

print("\n")

print("="*50)
print("STATISTICAL SUMMARY")
print("="*50)
print(df.describe())

print("\n")

print("="*50)
print("MISSING VALUES")
print("="*50)
print(df.isnull().sum())

# ==========================
# VISUALIZATION 1
# Exam Score Distribution
# ==========================

plt.figure(figsize=(8,5))

sns.histplot(df["Exam_Score"], bins=20, kde=True)

plt.title("Distribution of Exam Scores")
plt.xlabel("Exam Score")
plt.ylabel("Count")

plt.savefig("../images/exam_score_distribution.png")

plt.show()

# ==========================
# VISUALIZATION 2
# Correlation Heatmap
# ==========================

numeric_df = pd.get_dummies(df, drop_first=True)

plt.figure(figsize=(16,12))

sns.heatmap(
    numeric_df.corr(),
    cmap="coolwarm"
)

plt.title("Correlation Heatmap")

plt.savefig("../images/correlation_heatmap.png")

plt.show()

# ==========================
# VISUALIZATION 3
# Hours Studied vs Exam Score
# ==========================

plt.figure(figsize=(8,5))

sns.scatterplot(
    data=df,
    x="Hours_Studied",
    y="Exam_Score"
)

plt.title("Hours Studied vs Exam Score")

plt.savefig("../images/hours_vs_score.png")

plt.show()

# ==========================
# VISUALIZATION 4
# Attendance vs Exam Score
# ==========================

plt.figure(figsize=(8,5))

sns.scatterplot(
    data=df,
    x="Attendance",
    y="Exam_Score"
)

plt.title("Attendance vs Exam Score")

plt.savefig("../images/attendance_vs_score.png")

plt.show()

# ==========================
# VISUALIZATION 5
# Previous Scores vs Exam Score
# ==========================

plt.figure(figsize=(8,5))

sns.scatterplot(
    data=df,
    x="Previous_Scores",
    y="Exam_Score"
)

plt.title("Previous Scores vs Exam Score")

plt.savefig("../images/previous_scores.png")

plt.show()

# ==========================
# VISUALIZATION 6
# Gender vs Exam Score
# ==========================

plt.figure(figsize=(6,5))

sns.boxplot(
    data=df,
    x="Gender",
    y="Exam_Score"
)

plt.title("Gender vs Exam Score")

plt.savefig("../images/gender_score.png")

plt.show()

# ==========================
# VISUALIZATION 7
# Sleep Hours vs Exam Score
# ==========================

plt.figure(figsize=(8,5))

sns.scatterplot(
    data=df,
    x="Sleep_Hours",
    y="Exam_Score"
)

plt.title("Sleep Hours vs Exam Score")

plt.savefig("../images/sleep_hours.png")

plt.show()

print("\n")
print("="*50)
print("EDA COMPLETED SUCCESSFULLY")
print("="*50)
```
